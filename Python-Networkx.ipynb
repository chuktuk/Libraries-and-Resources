## Bipartite Graphs
- Graphs where there are two distinct groups of items that connect to each other but not within the groups
    - Customers/products
        - customers connect to products, but customers don't connect to each other and neither do products

```python
# bipartite graph
# the bipartite keyword will add the corresponding metadata to each node
G = nx.Graph()

numbers = range(8)
G.add_nodes_from(numbers, bipartite='customers')

letters = ['a', 'b']
G.add_nodes_from(letters, bipartite='products')

# add edges
elist = [(0, 'b', 5.0), (1, 'b', 3.0), (6, 'b', 1.0), (0, 'a', 7.3), (3, 'a', 2.0), (4, 'a', 1),
        (5, 'b', 6.0), (6, 'a', 3), (6, 'b', 1), (7, 'a', 2)]
G.add_weighted_edges_from(elist)

G.nodes(data=True)
```

```python
# filtering based on bipartite
# loop over nodes (n) and metadata dictionary (d) by specifying data=True
print('Customer List: {}'.format([n for n, d in G.nodes(data=True) if d['bipartite'] == 'customers']))
print('Product List: {}'.format([n for n, d in G.nodes(data=True) if d['bipartite'] == 'products']))
```

```python
# function for getting nodes of a specified partition (like bipartite)
# Define get_nodes_from_partition()
def get_nodes_from_partition(G, prop, partition):
    # Initialize an empty list for nodes to be returned
    nodes = []
    # Iterate over each node in the graph G
    for n in G.nodes():
        # Check that the node belongs to the particular partition
        if G.node[n][prop] == partition:
            # If so, append it to the list of nodes
            nodes.append(n)
    return nodes
```

```python
# call the function
get_nodes_from_partition(G, 'bipartite', 'products')
```

```python
# calculate degree centrality and display histogram
cust_nodes = get_nodes_from_partition(G, 'bipartite', 'customers')

# create the dictionary of degree centrality values
dcs = nx.degree_centrality(G)

# get individual customer dcs values using the nodes as indices
cust_dcs = [dcs[n] for n in cust_nodes]

# Plot the degree distribution of users_dcs
plt.yscale('log')
plt.hist(cust_dcs, bins=20)
plt.show()
```

## Bipartite graphs and recommendation systems

```python
# create the graph for simple GitHub users/repos
G = nx.Graph()

users = ['user1', 'user2', 'user3']
repos = ['repo1', 'repo2', 'repo3']

# populated the graph
G.add_nodes_from(users, bipartite='users')
G.add_nodes_from(repos, bipartite='repos')

# inpsect the graph
G.nodes(data=True)
```

```python
# create edges
edges = [('user1', 'repo2'), ('user2', 'repo2'), ('user3', 'repo2'), ('user3', 'repo1'), ('user2', 'repo1')]

# add the edges to the graph
G.add_edges_from(edges)

# inspect
G.edges()
```

```python
# visualize the graph
c = nv.CircosPlot(G, node_color='bipartite', node_grouping='bipartite', node_labels=True)
c.draw()
plt.show()
```

```python
# set comparison, must unpack the neighbors into a list
user1_neighbors = [n for n in G.neighbors('user1')]
user3_neighbors = [n for n in G.neighbors('user3')]

# set intersection
# creates a set (unique only items) for user1_neighbors
# intersection checks for any that appear in another list
set(user1_neighbors).intersection(user3_neighbors)
```

```python
# can also check differences for ones that don't appear

# items in user1 not appearing in user 3
print(set(user1_neighbors).difference(user3_neighbors))

# items in user3 not appearing in user 1
print(set(user3_neighbors).difference(user1_neighbors))
```

```python
# function to find overlap between two nodes
def shared_partition_nodes(G, node1, node2):
    # Check that the nodes belong to the same partition
    assert G.node[node1]['bipartite'] == G.node[node2]['bipartite']

    # Get neighbors of node 1: nbrs1
    nbrs1 = [n for n in G.neighbors(node1)]
    # Get neighbors of node 2: nbrs2
    nbrs2 = [n for n in G.neighbors(node2)]

    # Compute the overlap using set intersections
    overlap = set(nbrs1).intersection(nbrs2)
    return overlap
```

```python
# test the function
print(shared_partition_nodes(G, 'user1', 'user3'))
print(len(shared_partition_nodes(G, 'user1', 'user3')))
```

```python
# similarity function to test the similarity between two nodes from 1 bipartite to nodes in another bipartite
# this has a dependency on the shared_partition_nodes function above
def node_similarity(G, node1, node2, targ_nodes):
    # Check that the nodes belong to the 'users' partition
    assert G.node[node1]['bipartite'] == G.node[node2]['bipartite']

    # Get the set of nodes shared between the two users
    shared_nodes = shared_partition_nodes(G, node1, node2)

    # Return the fraction of nodes in the projects partition
    return len(shared_nodes) / len(targ_nodes)
```

```python
# function to find the most similar nodes to an input node using target nodes from another bipartite
from collections import defaultdict

def most_similar_nodes(G, node, nodes_list, node_bipartite, targ_nodes):
    
    # Data checks, to assert bipartites
    assert G.node[node]['bipartite'] == node_bipartite

    # Get other nodes from nodes_list partition
    nodes_list = set(nodes_list)
    nodes_list.remove(node)

    # Create the dictionary: similarities
    similarities = defaultdict(list)
    for n in nodes_list:
        similarity = node_similarity(G, node, n, targ_nodes)
        similarities[similarity].append(n)

    # Compute maximum similarity score: max_similarity
    max_similarity = max(similarities.keys())

    # Return list of users that share maximal similarity
    return similarities[max_similarity]
```

```python
# run the function to find the most similar user to user2

users = get_nodes_from_partition(G, 'bipartite', 'users')
repos = get_nodes_from_partition(G, 'bipartite', 'repos')

print(most_similar_nodes(G, 'user2', users, 'users', repos))
```

```python
# build recommendation function
def recommend_nodes(G, from_node, to_node):
    # Get the set of nodes that from_node has connected to
    from_set = set(G.neighbors(from_node))
    # Get the set of nodes that to_node has connected to
    to_set = set(G.neighbors(to_node))

    # Identify repositories that the from_user is connected to that the to_user is not connected to
    return from_set.difference(to_set)
```

```python
# test recommender function
print(recommend_nodes(G, 'user3', 'user1'))
```

### Projections of bipartite graphs
- Show relationships in one bipartite, based on their connections to the other bipartite
- A unipartite representation of a bipartite connectivity

```python
# building a projection

# isolate nodes of interest
user_nodes = [n for n in G.nodes() if G.node[n]['bipartite'] == 'users']

# call projected_graph function
# builds a graph containing only user_nodes, and projected edges
G_users = nx.bipartite.projected_graph(G, user_nodes)

# inspect graph
print(G_users.nodes())
print(G_users.edges())
```

```python
# view the new graph
# node_labels appears to do nothing for ArcPlots
a = nv.ArcPlot(G_users, node_labels=True)
a.draw()
plt.show()
```

### Matrix plots of bipartite graphs

```python
# using custom functions above to partition the data
users = get_nodes_from_partition(G, 'bipartite', 'users')
repos = get_nodes_from_partition(G, 'bipartite', 'repos')

# create the matrix
# done by multiplying the original matrix by it's transposition using matrix multiplication
mat = nx.bipartite.biadjacency_matrix(G, row_order=users, column_order=repos)

# the projection values will be the non-diagonal elements show which nodes are connected to which
# other nodes in the projection (connectivity matrix)
# compute the user-user matrix by multiplying the biad. matrix by its transpose
uum = mat @ mat.T
```

## Converting and storing a Graph as a dataframe
- Allows for pandas functions as well as saving to csv files

```python
import pandas as pd

# Initialize a list to store each edge as a record: nodelist
nodelist = []
for n, d in G.nodes(data=True):
    # nodeinfo stores one "record" of data as a dict
    nodeinfo = {'user': n} 
    
    # Update the nodeinfo dictionary 
    nodeinfo.update(d)
    
    # Append the nodeinfo to the node list
    nodelist.append(nodeinfo)
    

# Create a pandas DataFrame of the nodelist: node_df
node_df = pd.DataFrame(nodelist)
print(node_df)
```

```python
edgelist = []
for n1, n2, d in G.edges(data=True):
    # Initialize a dictionary that shows edge information: edgeinfo
    edgeinfo = {'node1':n1, 'node2':n2}
    
    # Update the edgeinfo data with the edge metadata
    edgeinfo.update(d)
    
    # Append the edgeinfo to the edgelist
    edgelist.append(edgeinfo)
    
# Create a pandas DataFrame of the edgelist: edge_df
edge_df = pd.DataFrame(edgelist)
print(edge_df)
```
